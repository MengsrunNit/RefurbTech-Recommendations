```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlencode

base_url = "https://www.ebay.com/sch/i.html"
params = {
    '_nkw': 'iPhone 15 Pro Max',
    '_sacat': '0',
    'LH_Sold': '1',
    'LH_Complete': '1',
    'LH_ItemCondition': '2010|2020|2030',
    '_ipg': '240',
}
url = f"{base_url}?{urlencode(params)}"
print(f"üîó Scraping URL: {url}\n")

headers = {'User-Agent': 'Mozilla/5.0'}
resp = requests.get(url, headers=headers)
resp.raise_for_status()

soup = BeautifulSoup(resp.text, 'html.parser')

# eBay's results list
results_ul = soup.select_one("ul.srp-results")  # more robust than exact class string

if not results_ul:
    print("‚ùå Could not find listings on the page (results <ul> not found).")
else:
    # Each result item
    listings = results_ul.select("li.s-item")[:4]  # first 4 items
    if not listings:
        print("‚ùå No items found under results <ul>. The page structure may have changed.")
    for li in listings:
        # Title + URL
        link = li.select_one("a.s-item__link")
        title_el = li.select_one(".s-item__title")
        title = title_el.get_text(strip=True) if title_el else None
        href = link.get("href") if link else None

        # Price
        price_el = li.select_one(".s-item__price")
        price = price_el.get_text(strip=True) if price_el else None

        # Condition (e.g., "Seller refurbished")
        condition_el = li.select_one(".SECONDARY_INFO")
        condition = condition_el.get_text(strip=True) if condition_el else None

        # Sold/completed info (date or status line)
        sold_info_el = li.select_one(".s-item__caption, .s-item__ended-date")
        sold_info = sold_info_el.get_text(strip=True) if sold_info_el else None

        print("Title:", title)
        print("URL:", href)
        print("Price:", price)
        print("Condition:", condition)
        print("Sold/Ended:", sold_info)
        print("-" * 60)
```
